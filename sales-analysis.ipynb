```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")
print(df.info())  # معلومات عن البيانات
print(df.describe())  # إحصائيات أساسية
```

```python
# تحليل المبيعات حسب المنتج
sales_per_product = df.groupby("Product")["Revenue"].sum()
sales_per_product.plot(kind="bar", title="إجمالي الإيرادات لكل منتج")
plt.show()
```

```python
# 1️⃣ شريط بياني - الإيرادات لكل منتج
plt.figure(figsize=(10, 5))
df.groupby("Product")["Revenue"].sum().plot(kind="bar", color="skyblue")
plt.title("إجمالي الإيرادات لكل منتج")
plt.xlabel("المنتج")
plt.ylabel("الإيرادات")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2️⃣ مخطط خطي - المبيعات عبر الزمن
df["Date"] = pd.to_datetime(df["Date"])
sales_per_day = df.groupby("Date")["Revenue"].sum()
plt.figure(figsize=(10, 5))
plt.plot(sales_per_day, marker="o", linestyle="-", color="green")
plt.title("الإيرادات اليومية")
plt.xlabel("التاريخ")
plt.ylabel("الإيرادات")
plt.grid()
plt.show()
```

```python
# 3️⃣ مخطط دائري - توزيع المبيعات حسب الفئات
category_sales = df.groupby("Category")["Revenue"].sum()
plt.figure(figsize=(7, 7))
plt.pie(category_sales, labels=category_sales.index, autopct="%1.1f%%", colors=["blue", "red", "green"])
plt.title("توزيع الإيرادات حسب الفئات")
plt.show()
```
